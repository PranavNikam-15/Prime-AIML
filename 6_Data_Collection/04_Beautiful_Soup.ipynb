```python
# BeautifulSoup

from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data1.html", "r", encoding="utf-8") as f:
    html_content = f.read()


soup = BeautifulSoup(html_content, "lxml")
print(soup.title.text)
```

```python
# find(): returns the first occurrence of the tag
print(soup.find("h3"))


# find_all(): returns a list of all occurrences of the tag
all_h3 = soup.find_all("h3")  # Find all <h3> tags
print("\nNumber of H3 tags :", len(all_h3))
```

```python
for h3 in all_h3:
    
    # Country name
    name = h3.get_text(strip=True)
    
    # find_next(): find the next <div> after <h3>
    next_div = h3.find_next("div")
    
    # select() to get the population span inside that div
    population_span = next_div.select("span.country-population")[0]
    population = population_span.get_text(strip=True)
    
    # find_parent(): get the parent div of the <h3> tag
    parent_div = h3.find_parent("div")
    
    print("Country:", name)
    print("Population:", population)
    print("Parent Div class:", parent_div.get("class"))
    print("------------------------------------------")
```
