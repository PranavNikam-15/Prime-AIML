```python
# Web Scraping :
# - Web scraping means automatically collecting data from websites using a program instead of copying it manually.

import requests
res = requests.get("https://www.scrapethissite.com/pages/simple/")
```

```python
# Check whether the request was successful (HTTP Status Code 200 = OK)

if res.status_code == 200:
    # res.text : returns response data as string (decoded text)
    # res.content : returns response data as raw bytes
    # res.headers : return response headers (metadata from server)
    print(res.headers)
else:
    print(res.status_code)
```

```python
# Save the scraped HTML content into a file

with open("scraped_data/data1.html", "w", encoding="utf-8") as f:
    f.write(res.text)
```

```python
import time

# Headers:
# - Some websites block requests that do not look like a real browser
# - User-Agent tells the server that the request is coming from a browser

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

try:
    res = requests.get(
        "https://www.scrapethissite.com/pages/simple/",
        headers=headers
    )
    
    # Raise exception if request failed (404, 500)
    res.raise_for_status()

except requests.exceptions.RequestException as e:
    print("Error:", e)


# Add delay:
# - Prevents sending too many requests at once
# - Helps avoid getting blocked by the website
time.sleep(2)   # sleep for 2 seconds

print("Done!")
```
