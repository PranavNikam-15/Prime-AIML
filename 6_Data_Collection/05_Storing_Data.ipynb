```python
# BeautifulSoup

from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data1.html", "r", encoding="utf-8") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
all_h3 = soup.find_all("h3")

all_countries = []

for h3 in all_h3:
    
    # Country name
    name = h3.get_text(strip=True)
    
    # find_next(): find the next <div> after <h3>
    next_div = h3.find_next("div")
    
    # select() to get the population span inside that div
    population_span = next_div.select("span.country-population")[0]
    population = population_span.get_text(strip=True)

    all_countries.append([name, population])
```

```python
import pandas as pd

df = pd.DataFrame(all_countries, columns=["Country", "Population"])
df
```

```python
# Store data into csv file
df.to_csv("country_data.csv")
```
