```python
# Start by loading the TITANIC data set

import pandas as pd

# load dataset
df = pd.read_csv('Titanic.csv')
```

```python
# Display only columns : Name, Sex, Age, Fare, Survived

df[["Name", "Sex", "Age", "Fare", "Survived"]]
```

```python
# Select passengers who : 1) Are female  2) Fare > 30

df.query("Sex == 'female' ")["Name"]
df[df["Sex"] == "female"]


df.query("Fare > 30")
```

```python
# Group by Pclass and compute :  

# Average Fare
df.groupby("Pclass")["Fare"].aggregate("mean")

# Average Age
df.groupby("Pclass")["Age"].mean()
```
