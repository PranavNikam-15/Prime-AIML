```python
import pandas as pd
```

```python
# Merging & Joining

# Customers DataFrame
df_customers= pd.DataFrame({ 
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
}) 


# Orders DataFrame
df_orders= pd.DataFrame({ 
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount":[250, 120, 300, 180]
})
```

```python
# Merging & Joining : 

# Inner Join: Returns only rows with matching 'customer_id' in both DataFrames
pd.merge(df_customers, df_orders, on="customer_id")
```

```python
# Merging & Joining : 

# Left Join: Returns all rows from df_customers and matching rows from df_orders
# Non-matching rows in df_orders will have NaN values
pd.merge(df_customers, df_orders, on="customer_id", how="left")
```

```python
# Merging & Joining : 

# Right Join: Returns all rows from df_orders and matching rows from df_customers
# Non-matching rows in df_customers will have NaN values
pd.merge(df_customers, df_orders, on="customer_id", how="right")
```

```python
# Merging & Joining : 

# Outer Join: Returns all rows from both DataFrames
# Non-matching rows will have NaN values in columns where no match is found
pd.merge(df_customers, df_orders, on="customer_id", how="outer")
```
