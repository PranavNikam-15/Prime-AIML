```python
import pandas as pd

# import data
df = pd.read_csv('raw_data.csv')
```

```python
# Transforming Data

# sort_values() : # Sort rows based on Salary column (ascending by default)
sorted_data = df.sort_values("income", ascending=False)
sorted_data

# sorted_data.sort_index()

# reset_index() : creates a new sequential index
sorted_data.reset_index()

# reset_index(drop=True) : removes old index completely
sorted_data.reset_index(drop=True)
```

```python
# Transforming Data : Ranking

# rank() : Assign rank based on Salary (lowest salary - rank 1)
sorted_data["Ranking"] = sorted_data["income"].rank()

# ascending=False, highest salary - rank 1
# dense ranking (no gaps in ranks)
sorted_data["Ranking"] = sorted_data["income"].rank(ascending=False, method="dense")
sorted_data
```

```python
# Transforming Data

new_data = df.copy()

# Changing column names

# new_data.columns = ["Id",	"Name",	"Age", "Country", "Gender", "Salary"]
new_data.columns = new_data.columns.str.capitalize()

# rename() :
# `rename(columns={...})` allows changing selected column names
new_data = new_data.rename(columns={"Income": "Salary"})

# rename() :
# `rename(index={...})` allows changing selected row labels
new_data = new_data.rename(index={1: "First"})

new_data
```
