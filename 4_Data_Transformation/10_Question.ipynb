```python
# Start by loading the IRIS data set

import pandas as pd

# load dataset
df = pd.read_csv('IRIS.csv')
```

```python
# Display the following information: 

# First 10 rows
df.head(10)

# Shape & Datatypes 
df.shape
df.dtypes

# Summary statistics (mean,std,min,max)
df.describe()
```

```python
# Select only those rows where :


# petal_length > 4.5
df[df["petal_length"] > 4.5]
df.query("petal_length > 4.5")


# species="Iris-virginica"
df.query("species == 'Iris-virginica'")
```

```python
# Group by species and compute :

# Average sepal_length 
df.groupby("species")["sepal_length"].agg("mean")

# Maximum petal_width
df.groupby("species")["petal_width"].agg("max")

# Standard deviation of sepal_width
df.groupby("species")["sepal_width"].agg("std")
```

```python
# Create a new column : "petal_ratio" = petal_length/petal_width

new_df = df.assign(petal_ratio = df["petal_length"]/df["petal_width"])
new_df.head(10)

# For each species, find the average petal_ratio
new_df.groupby("species")["petal_ratio"].agg("mean")
```
