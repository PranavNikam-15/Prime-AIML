```python
import pandas as pd
import numpy as np
# import data
df = pd.read_csv('raw_data.csv')
new_data = df.copy()
```

```python
# Feature Engineering

# apply() : Applies a function to each element, row or column of a series or DataFrame
new_data["tax"] = new_data["income"].apply(lambda x : "20%" if x>= 60000 else "10%")
new_data
```

```python
# Feature Engineering

# map() : Applies a function or mapping element-wise on a Series only
gender_map = { "Male" : "M", "Female" : "F", "Unknown" : "U" }
new_data["gender"] = new_data["gender"].map(gender_map)
new_data.loc[0:5, ["name", "age", "gender"]]
```

```python
# Feature Engineering

# assign() : Creates or modifies columns in a DataFrame and returns a new DataFrame

# Add new column
new_data = new_data.assign(new_income = new_data["income"]*1.1)
new_data.loc[0:5]
```

```python
# Feature Engineering

# replace() : Used to replace specific values in a Series or DataFrame
# Can replace using values, lists, or dictionaries

new_data["name"] = new_data["name"].replace(np.nan, "Not Available")
new_data.loc[5:7, "name"]

# Replace values across entire DataFrame
# new_data.replace(np.nan, "Not Available")
```
