```python
import pandas as pd
```

```python
# DataFrames

df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Pranav", "Buddy", "Ketan"]
})

df2 = pd.DataFrame({
    "ID": [4, 5, 6],
    "Name": ["Shivraj", "Viraj", "Ruchir"]
})

df3 = pd.DataFrame({
    "Age": [23, 22, 30],
    "City": ["Phaltan", "Pune", "Delhi"]
})
```

```python
# vertical Concatenation

vertical_concat_data = pd.concat([df1, df2], axis=0)
vertical_concat_data

vertical_concat_data = pd.concat([df1, df2], axis=0, ignore_index=True)
vertical_concat_data

# ignore_index=True : Reindexes the result after vertical concatenation
```

```python
# Horizontal Concatenation
# Stack DataFrames horizontally (adds columns)

# Rows must match
horizontal_concat = pd.concat([df1, df3], axis=1)
print(horizontal_concat)
```
