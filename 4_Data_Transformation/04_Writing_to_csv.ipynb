```python
import pandas as pd

# import data
df = pd.read_csv('raw_data.csv')
```

```python
new_data = df.copy()

# Remove duplicate rows 
new_data.drop_duplicates(inplace=True)

# Handle missing values
age_mean = round(new_data["age"].mean())
new_data["age"] = new_data["age"].fillna(age_mean)

new_data["income"] = new_data["income"].ffill()

common_country = new_data["country"].mode()[0]
new_data["country"] = new_data["country"].fillna(common_country)

new_data["gender"] = new_data["gender"].fillna("Unknown")
new_data["name"] = new_data["name"].fillna("Not Available")

# Sort data

new_data = new_data.sort_values("income")
new_data = new_data.reset_index(drop=True)
new_data
```

```python
# Write into csv file

new_data.to_csv("Cleaned_raw_data.csv")
```
