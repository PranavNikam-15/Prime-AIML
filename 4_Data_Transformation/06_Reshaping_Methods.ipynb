```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 18, 22] 
}) 

df
```

```python
# melt() : Converts wide-format data into long-format data

# id_vars   : Columns to keep as identifiers
# value_vars: Columns to unpivot
# var_name  : Name of the column that stores former column names
# value_name: Name of the column that stores values

melted_data = df.melt(
    id_vars=["country", "year"],
    value_vars=["sales", "profit"],
    var_name="metrics",
    value_name="value"
)
```

```python
# pivot() : Converts long-format data back into wide-format

# index   : Columns to use as row index
# columns : Column whose unique values become new columns
# values  : Column providing the values for the new table

original_data = melted_data.pivot(
    index = ["country", "year"],
    columns = "metrics",
    values="value"   
)

original_data
```
