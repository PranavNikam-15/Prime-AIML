```python
import pandas as pd
```

```python
s = pd.Series([10, 20, 30, 40, 50])

print(s.sum())
print(s.mean())
print(s.median())
print(s.max())
print(s.min())
print(s.std())
print("----------------------")
print(s.describe())
```

```python
# Handling missing values

s = pd.Series([10, None, 30, None, 50])

print(s.isnull())
print(s.notnull())
print(s.fillna(0))
print(s.dropna())
```
