```python
import pandas as pd
```

```python
'''
A DataFrame is a 2-dimensional, labeled data structure with :
- Rows (index)
- Columns (labels)
- Each column can have a different data type

'''

# Dataframe from dictionary
data = {
    "Name" : ["Pranav", "Buddy", "Shivraj"],
    "Age" : [23, 22, 22],
    "GPA" : [8.4, 9.1, 7.9]
}

df = pd.DataFrame(data)
print(df)
```

```python
# DataFrame from list of lists

data = [ ["Pranav", 23, "Phaltan"], ["Buddy", 22, "Pune"], ["Shivraj", 21, "Nashik"] ]

df = pd.DataFrame(data, columns=["Name", "Age", "City"])
print(df)
```

```python
# Accessing DataFrame

print(df["Name"])
df[["Name", "Age"]]
```
