```python
import pandas as pd
```

```python
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])

print(s1 + s2)
print(s2 - s1)
print(s1 * 2)
print(s2 / 10)
```

```python
# Operations with different indexes

s1 = pd.Series([1, 2, 3], index=["a", "b", "c"])
s2 = pd.Series([4, 5, 6], index=["b", "c", "d"])

print(s1 + s2)
```

```python
# Comparision Operations 

s = pd.Series([15, 20, 25, 30, 35], index=["a", "b", "c", "d", "e"])

print(s > 25) # true/false series
print(s == 15)
```
