```python
import pandas as pd
```

```python
# Series is a one-dimensional labeled array that can hold data of any type (integer, float, string, Python objects)

s = pd.Series([10, 20, 30, 40, 50])
print(s)
```

```python
# Accessing series using index

print(s[0])
print(s[4])

# Series with labelled names
s = pd.Series([10, 20, 30, 40, 50], index=["Pranav", "Buddy", "Ketan", "Nishad", "Ruchir"])
print(s)

# Accessing series using label names
print(s["Pranav"])
print(s["Ruchir"])

print(s.loc["Ketan"])  # label-based
print(s.iloc[2])   # position-based (index)
```

```python
# Series using dictionary
data = { "Pranav": 100, "Buddy": 200, "Rohan": 300, "John": 400 }
s = pd.Series(data)
print(s)

# Series using scalar
s = pd.Series(15, index=["a", "b", "c"])
print(s)
```
