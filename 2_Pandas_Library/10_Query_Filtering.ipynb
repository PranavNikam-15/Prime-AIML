```python
import pandas as pd

# import csv file
df = pd.read_csv('globalAirQuality.csv')
```

```python
# query

df.query("aqi > 100")
df.query("aqi > 100 & temperature > 30")

aqi_val = 195
df.query("aqi >= @aqi_val & temperature > 30")[["city", "aqi"]]
```

```python
# query

# Select rows where city is Mumbai
df.query("city == 'Mumbai'")

# list of cities
cities = ["Mumbai", "Delhi"]
# Select rows where the "city" column value is present in the cities list
df.query("city in @cities")
```
