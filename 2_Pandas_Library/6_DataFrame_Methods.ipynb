```python
import pandas as pd
```

```python
df = pd.read_csv('employee_data.csv')
df
```

```python
# head(n): Returns the first n rows of the DataFrame
# By default, it returns the first 5 rows
df.head()
```

```python
# tail(n): Returns the last n rows of the DataFrame
# By default, it returns the last 5 rows
df.tail() 
```

```python
# sample(n): Returns n random rows from the DataFrame
# By default, it returns 1 random row
df.sample()
```

```python
# info(): Displays a concise summary of the DataFrame
# Includes column names, number of non-null values, and data types
df.info()
```

```python
# describe(): Generates descriptive statistics for numerical columns
# Includes count, mean, standard deviation, min, max, and quartiles
df.describe()
```

```python
# dtypes: Returns the data type of each column
df.dtypes
```

```python
# shape: Returns the dimensions of the DataFrame
# Format: (number of rows, number of columns)
df.shape
```

```python
# columns: Returns an index object containing column names
df.columns
```

```python
# nunique(): Returns the number of unique values in each column
df.nunique()
```
