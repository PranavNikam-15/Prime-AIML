```python
import pandas as pd
```

```python
df = pd.read_csv('globalAirQuality.csv')
```

```python
# selecting data

# columns
df["country"]
df[["country","city", "aqi"]]
```

```python
# selecting data
# rows

# loc : selects rows by label/index name
df.loc[0]
df.loc[0:6:2] # start index : end index (enclusive) : step


# iloc : selects rows by integer positions (indexing)
df.iloc[5]
df.iloc[0:6:2] # start index : end index (exclusive) : step

# loc : label-based selection (includes the last index in slicing)
# iloc : position-based selection (excludes the last index in slicing)
```

```python
# selecting data
# cells - row & col

df.loc[0, "city"]
df.loc[0:3, ["city", "country"]]

df.columns

#df.iloc[0:5, [0, 1, 2]]
df.iloc[0:6, 1:5]
```

```python
# selecting data
# cell - single scalar value

df.at[0, "city"]
df.iat[0, 2]
```
