```python
import pandas as pd

# import data
df = pd.read_csv('globalAirQuality.csv')
```

```python
# filter data

# AQI is greater than 100
df[df["aqi"] > 100]

# AQI is greater than 100 &  Temperature is less than 30
df[ (df["aqi"] > 100) & (df["temperature"] < 30) ]

# From the filtered rows (AQI > 100 and temperature > 30)
# select only the "city" and "aqi" columns
aqi_data = df[ (df["aqi"] > 100) & (df["temperature"] > 30)] [["city", "aqi"]]
aqi_data

# Accessing rows using iloc

# iloc works with POSITION (integer index)
# iloc[0] returns the FIRST row of aqi_data
aqi_data.iloc[0]

# Accessing rows using loc

# loc works with INDEX LABELS
# loc[6] returns the row whose index label is 6
aqi_data.loc[6]
```
