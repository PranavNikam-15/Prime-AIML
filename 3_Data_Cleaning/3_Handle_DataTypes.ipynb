```python
import pandas as pd

# import data
df = pd.read_csv('raw_data.csv')
data = df.copy()

data
```

```python
# Cleaning Data : Handling Data Types

age_mean = data["age"].mean()
data["age"] = data["age"].fillna(age_mean)


# conversion of data type : float to int
print(data["age"].dtypes)
data["age"] = data["age"].astype("int64")
print(data["age"].dtypes)
```

```python
aqi_data = pd.read_csv('globalAirQuality.csv')
print(aqi_data["timestamp"].dtype)

# conversion of data type : object to datetime
aqi_data["timestamp"] = pd.to_datetime(aqi_data["timestamp"])
print(aqi_data["timestamp"].dtype)
```
