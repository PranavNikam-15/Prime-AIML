```python
import pandas as pd

# import data
df = pd.read_csv('raw_data.csv')
```

```python
# Cleaning Data : Handling Strings

gender = df["gender"]

# Convert all strings to lowercase
gender.str.lower()

# Convert all strings to uppercase
gender.str.upper()

# Capitalize the first letter of each string
gender.str.capitalize()
```

```python
# Cleaning Data : Handling Strings

name = df["name"]

# Split each name by space
name.str.split(" ")

# "pranav@gmail.com".split("@")  ==> ['pranav', 'gmail.com']
```

```python
country = df["country"]

# Check if 'india' exists in the country column
country.str.contains("india", case=False, na=False)

# case : Controls case sensitivity of the search
# na : Controls what to return for missing values (NaN)
```
