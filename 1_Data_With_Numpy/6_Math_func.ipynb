```python
import numpy as np
```

```python
# Sample array
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Minimum and Maximum values
arr_min = arr.min()          # Minimum value
arr_max = arr.max()          # Maximum value
print("Min:", arr_min)
print("Max:", arr_max)

# Indices of Minimum and Maximum values
arr_min_index = arr.argmin()  # Index of minimum value
arr_max_index = arr.argmax()  # Index of maximum value
print("Min Index:", arr_min_index)
print("Max Index:", arr_max_index)

# Sum and Product of all elements
arr_sum = np.sum(arr)         # Sum of all elements
arr_prod = np.prod(arr)       # Product of all elements
print("Total Sum:", arr_sum)
print("Total Product:", arr_prod)

# Mean, Median
arr_mean = np.mean(arr)       # Average value
arr_median = np.median(arr)   # Middle value
print("Mean:", arr_mean)
print("Median:", arr_median)

# Variance and Standard Deviation
arr_var = np.var(arr)         # Variance
arr_std = np.std(arr)         # Standard deviation
print("Variance:", arr_var)
print("Standard Deviation:", arr_std)
```

```python
# Power Functions

arr = np.array([1, 2, 3, 4, 5])

# Square of each element
sq_arr = np.square(arr)
print("Square Array:", sq_arr)

# Square root of each element
sqrt_arr = np.sqrt(arr)
print("Square-Root Array:", sqrt_arr)

# Power of each element
power_arr = np.power(arr, 3)
print("Power Array:", power_arr)
```

```python
# Log Functions

arr = np.array([1, 2, 3, 4, 5])

# Natural logarithm (base e)
arr_log = np.log(arr)
print("log:", arr_log)

# Logarithm base 2
arr_log2 = np.log2(arr)
print("log2:", arr_log2)

# Logarithm base 10
arr_log10 = np.log10(arr)
print("log10:", arr_log10)

# Exponential function (e^x)
arr_exp = np.exp(arr)
print("exp:", arr_exp)
```

```python
# Rounding Functions

# round() : rounds to nearest integer
print(np.round(3.14))   # 3.0
print(np.round(3.84))   # 4.0

# ceil() : rounds up to the next integer
print(np.ceil(3.14))    # 4.0

# floor() : rounds down to the previous integer
print(np.floor(3.94))   # 3.0

# trunc() : removes decimal part
print(np.trunc(22.9191))  # 22.0
```
