```python
import time
```

```python
#  execution performance
size = 10_000_000

# python List
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()

print(f"Python List : {end-start} seconds")

# numpy array
np_array = np.array(py_list) 
start = time.time()

# vectorization
sq_array = np_array ** 2
end = time.time()
print(f"Numpy Array : {end-start} seconds")
```

```python
import sys

print(f"Python list size : {sys.getsizeof(py_list) * len(py_list)} Bytes")
print(f"Numpy Array size : {np_array.nbytes} Bytes")
```
